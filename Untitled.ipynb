```python
import pandas as pd
```

```python
products  = pd.read_csv('./data/Weekly_products.csv',low_memory=False)
```

```python
nulls = pd.read_csv('./data/Weekly_NA_Report.csv',low_memory=False)
```

```python
products.drop(columns='attribute_id',inplace=True)
```

```python
products.drop_duplicates(keep='first',inplace=True)
```

```python
merge = nulls.merge(products,on='product_id')
```

```python
cols = ['customer_id_x','bucket_id','attribute_name','Total_NAs','curation_tasks_count','active']
```

```python
merge[cols].groupby(by=['customer_id_x','bucket_id','attribute_name']).sum()
```
